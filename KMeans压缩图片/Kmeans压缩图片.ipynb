```python
import sklearn.datasets as datasets
import matplotlib.pyplot as plt 
from  sklearn.cluster import KMeans
```

```python
image = datasets.load_sample_image('china.jpg')
```

```python
plt.imshow(image)
```

```python
x_train = image.reshape((-1,3))
x_train.shape
```

```python
# 使用64个质点来对数据进行聚类
kmeans = KMeans(64)
kmeans.fit(x_train)
y_ = kmeans.predict(x_train)
```

```python
cluster_centers = kmeans.cluster_centers_
cluster_centers
```

```python
new_china = cluster_centers[y_]/256
plt.imshow(new_china.reshape((427,640,3)))
```

```python
plt.imsave('./new_China.png',new_china.reshape((427,640,3)))
```
