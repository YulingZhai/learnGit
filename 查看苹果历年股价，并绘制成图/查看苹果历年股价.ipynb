```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
```

```python
apple_price = pd.read_csv('./AAPL .csv')
apple_price
```

```python
# 观察股价的各种数据类型
apple_price.dtypes
```

```python
# 将Date的类型转换为时间类型
# 将Date这一列进行转换，然后赋值给date这一列
apple_price['Date'] = pd.to_datetime(apple_price['Date']) 
```

```python
apple_price.dtypes
```

```python
apple_price.head()
```

```python
# 将Date这一列设置维索引
apple_price.set_index('Date',inplace=True)
```

```python
apple_price.head()
```

```python
# 有于市值一项太大，将其删掉、绘制整个图像
apple_Adj = apple_price.drop(axis = 1, labels='Volume').plot()
fig = apple_Adj.get_figure()
fig.set_size_inches(12,9)
```
