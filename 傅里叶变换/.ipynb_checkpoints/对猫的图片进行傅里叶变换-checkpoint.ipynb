```python
import numpy as np
from numpy.fft import fft, ifft
import matplotlib.pyplot  as plt
```

第一步读取数据

```python
cat = Image.open('./cat.jpg')
plt.imshow(cat)
```

```python
type(cat)
```

```python
# 从cat图片文件中获取数据
cat_data = np.fromstring(cat.tobytes(),np.int8)
cat_data.shape
```

```python
cat_fft = fft(cat_data)
# 实数加上虚数部分
cat_fft
```

```python
# 滤除低频的波
# 保留高频的波--->保留了图片中变化较大的波。
cond = np.abs(cat_fft) < 1e5
cat_fft[cond] = 0
```

```python
使用傅里叶反转
```

```python
cat_ifft = ifft(cat_fft)
cat_ifft
```

```python
# 取实数部分
cat_real = np.real(cat_ifft)
cat_real
```

```python
cat_result = np.int8(cat_real)
cat_result
```

```python
# 将过滤掉低频波的数据，转换成图片
cat_outlook = Image.frombytes(mode='RGB', size=(730,456),data=cat_result)
```

```python
# 显示最后形成轮廓的猫。
cat_outlook
```
